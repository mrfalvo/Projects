### Algorithim for Analysis of Major G10 Currencies ###

```python
class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'


def Statfinder(Product_data):

    Product = Product_data.iloc[:, [0,1,2,3,4]]
    Product.columns = ["Date","Close", "Open", "High", "Low"] #rename column name "Price" to "Close"
    Product = Product.iloc[5:, :] #Clean data so that the most recent data starts from a full week

    Weekly_High_Day = []
    Weekly_Low_Day = []

    WeekIndex_num1 = -5
    WeekIndex_num2 = 0
    
    Fri_Close_Criteria = -1

    def remove_neg(num):
        num = str(num)
        num = float(num.replace("-", ""))
        return num

    
    Test1 = []
    Test2 = []

    MonW = []
    TuesW = []
    WedW = []
    

    for i in range(980):
    
        WeekIndex_num1 += 5
        WeekIndex_num2 += 5
    
        a = (Product.iloc[WeekIndex_num1:WeekIndex_num2, :])
        
    
        Weekly_Highs = pd.Series(data=[a.iloc[0,3], a.iloc[1,3], a.iloc[2,3], a.iloc[3,3], a.iloc[4,3]], 
                        index=['Fri', 'Thurs', 'Wed', 'Tues', 'Mon'])
        Weekly_Lows = pd.Series(data=[a.iloc[0,4], a.iloc[1,4], a.iloc[2,4], a.iloc[3,4], a.iloc[4,4]],
                       index=['Fri', 'Thurs', 'Wed', 'Tues', 'Mon'])

        Weekly_Close = pd.Series(data=[a.iloc[1,1], a.iloc[2,1], a.iloc[3,1], a.iloc[4,1]], 
            index=['Thurs', 'Wed', 'Tues', 'Mon'])
        
        Weekly_Open = pd.Series(data=[a.iloc[0,2], a.iloc[1,2], a.iloc[2,2], a.iloc[3,2], a.iloc[4,2]], 
            index=['Fri', 'Thurs', 'Wed', 'Tues', 'Mon'])
        
        Day_High = Weekly_Highs.idxmax()

        Day_Low = Weekly_Lows.idxmin()
        
        Day_Close_Low = Weekly_Close.idxmin()
        
        Day_Close_High = Weekly_Close.idxmax()

        Weekly_High_Day.append(Day_High)
        Weekly_Low_Day.append(Day_Low)
        
        if Day_High == "Thurs" and Day_Close_Low == "Thurs":
            x = a.iloc[0,0]
            Test1.append(x)
            
        if Day_Low == "Thurs" and Day_Close_High == "Thurs":
            p = a.iloc[0,0]
            Test2.append(p)
            
    
    
        Wed = remove_neg(((float(a.iloc[2,1]) - float(a.iloc[0,2])) / float(a.iloc[0, 2])) * 100) 
        Tues = remove_neg(((float(a.iloc[3,1]) - float(a.iloc[1,2])) / float(a.iloc[1, 2])) * 100) 
        Mon = remove_neg(((float(a.iloc[4,1]) - float(a.iloc[2,2])) / float(a.iloc[2, 2])) * 100)
    

        if Mon > Tues and Mon > Wed:
            MonW.append(Mon)
    
        elif Tues > Mon and Tues > Wed:
            TuesW.append(Tues)
    
        elif Wed > Mon and Wed > Tues:
            WedW.append(Wed)
    
    
    print(color.BOLD + "Stats for high/low of the week occurances:" + color.END)
    print("\n")
    print("Mon:",((Weekly_High_Day.count("Mon") + Weekly_Low_Day.count("Mon")) / 1960) * 100)
    print("Tues:",((Weekly_High_Day.count("Tues") + Weekly_Low_Day.count("Tues")) / 1960) * 100)
    print("Wed:",((Weekly_High_Day.count("Wed") + Weekly_Low_Day.count("Wed")) / 1960) * 100)
    print("Thurs:",((Weekly_High_Day.count("Thurs") + Weekly_Low_Day.count("Thurs")) / 1960) * 100)
    print("Fri:",((Weekly_High_Day.count("Fri") + Weekly_Low_Day.count("Fri")) / 1960) * 100)
    
    print("\n")

    print(color.BOLD + "Stats for largets three day one directional move of the week:" + color.END)
    print("\n")
    print("Mon-Wed Occurances:",len(MonW), "Avg Move:", statistics.median(MonW))
    print("Tues-Thurs Occurances:",len(TuesW), "Avg Move:", statistics.median(TuesW))
    print("Wed-Fri Occurances:",len(WedW), "Avg Move:", statistics.median(WedW))
    
    print("\n")
    print(color.BOLD + "List of Occurences where Thursday was the highest/lowest price so far in the week, while closing the day at the lowest/highest close of the week:" + color.END)
    print("\n")
    print(Test2)
    print(Test1)
```
